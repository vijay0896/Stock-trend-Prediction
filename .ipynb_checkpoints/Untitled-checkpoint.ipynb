```python
import yfinance as yf
import mplfinance as mpf

# Define the ticker symbol
ticker_symbol = "AAPL"

# Fetch stock data
stock_data = yf.download(ticker_symbol, start="2023-10-01", end="2024-01-01")

# Plot the candlestick chart
mpf.plot(stock_data, type='candle', style='yahoo', volume=True)
```
